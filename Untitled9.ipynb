```python
plt.figure(figsize = (14,8))
plt.plot(x.index,x,marker = 'o',markersize = 10,mfc = 'r',linestyle = '--')
plt.xlabel('Car Brand',size = 14,weight = 'bold')
plt.ylabel('Avg. Price',size = 14,weight = 'bold')
plt.title('Avg.-price VS Car Brand',size = 16,weight = 'bold')
plt.xticks(rotation = 90)
plt.show()
```
